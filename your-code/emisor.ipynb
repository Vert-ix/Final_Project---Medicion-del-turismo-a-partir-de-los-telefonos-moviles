```python
df.columns= ['Fecha', 'Turistas','Pernoctaciones','Duracion media de los viajes','Zona']
df.head(50)
```

```python
# Renombrar columnas

new_cols = [c.lower().rstrip().lstrip().replace(' ', '_') for c in df.columns]
new_cols
```

```python
dict(zip(df.columns, new_cols))
```

```python
df = df.rename(columns={k:v for k,v in zip(df.columns, new_cols)})
df.head(50)
```

```python
# Rellenar valores nulos con 0

df.fillna({'turistas': 0, 'name': 0}, inplace=True)
df.fillna({'pernoctaciones': 0, 'name': 0}, inplace=True)
df.fillna({'duracion_media_de_los_viajes': 0, 'name': 0}, inplace=True)
df.head()
```

```python
df.turistas = df.turistas.replace({".": 0," ": 0})
df.pernoctaciones = df.pernoctaciones.replace({".": 0," ": 0})
df.duracion_media_de_los_viajes = df.duracion_media_de_los_viajes.replace({".": 0," ": 0})
```

```python
df = df.astype({'turistas': 'int32', 'pernoctaciones': 'int32', 'duracion_media_de_los_viajes': 'int32'})
```

```python
cambio_letras={'á': 'a', 'é': 'e', 'í': 'i', 'ó':'o', 'ú': 'u'}

def quitar_acentos(palabra):
    
    resultado=''
    
    palabra=palabra.lower()
    
    for letra in palabra:
        
        if letra in cambio_letras.keys():
            
            resultado+=cambio_letras[letra]
            
        else:
            resultado+=letra
            
    return resultado.capitalize()
            

df['zona']=df['zona'].apply(quitar_acentos)

df['zona'].head()
```

```python
# Cambiar columnas a valores numéricos

df['turistas']=pd.to_numeric(df['turistas'],downcast='integer')
df['pernoctaciones']=pd.to_numeric(df['pernoctaciones'],downcast='integer')
df['duracion_media_de_los_viajes']=pd.to_numeric(df['duracion_media_de_los_viajes'],downcast='integer')
```
