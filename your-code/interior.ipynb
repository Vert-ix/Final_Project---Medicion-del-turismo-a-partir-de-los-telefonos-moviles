```python
df = pd.read_excel('../interior/municipio_origen_destino.xlsx')
df
```

```python
cambio_letras={'á': 'a', 'é': 'e', 'í': 'i', 'ó':'o', 'ú': 'u'}

def quitar_acentos(palabra):
    
    resultado=''
    
    palabra=palabra.lower()
    
    for letra in palabra:
        
        if letra in cambio_letras.keys():
            
            resultado+=cambio_letras[letra]
            
        else:
            resultado+=letra
            
    return resultado.capitalize()
            

df['mun_orig']=df['mun_orig'].apply(quitar_acentos)

df['mun_orig'].head()
```

```python
df['dest']=df['dest'].apply(quitar_acentos)
```

```python
df['prov_orig']=df['prov_orig'].apply(quitar_acentos)
```

```python
df['prov_dest']=df['prov_dest'].apply(quitar_acentos)
```

```python
df.replace({'prov_orig_cod':dic}, inplace=True)
```
