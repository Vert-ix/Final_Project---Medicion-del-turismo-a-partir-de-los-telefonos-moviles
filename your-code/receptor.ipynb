```python
cambio_letras={'á': 'a', 'é': 'e', 'í': 'i', 'ó':'o', 'ú': 'u'}

def quitar_acentos(palabra):
    
    resultado=''
    
    palabra=palabra.lower()
    
    for letra in palabra:
        
        if letra in cambio_letras.keys():
            
            resultado+=cambio_letras[letra]
            
        else:
            resultado+=letra
            
    return resultado.capitalize()
            

df['zona']=df['zona'].apply(quitar_acentos)

df['zona'].head()
```

```python
# Cambiar columnas a valores numéricos

df['turistas']=pd.to_numeric(df['turistas'],downcast='integer')
df['pernoctaciones']=pd.to_numeric(df['pernoctaciones'],downcast='integer')
df['duracion_media_de_los_viajes']=pd.to_numeric(df['duracion_media_de_los_viajes'],downcast='integer')
```
